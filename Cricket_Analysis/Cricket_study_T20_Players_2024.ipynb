```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import warnings
#warnings.filterwarning('ignore')
```

```python
data1 = pd.read_html('https://www.espncricinfo.com/records/year/averages-batting/2024-2024/twenty20-matches-6')
```

```python
df = data1[0]
df.drop(df.columns[1], axis = 1, inplace = True)
df
```

```python
#Changing the data in table to csv
df.to_csv('Average_batting.csv')
```

```python
df.head()
```

```python
#Top player by runs
Top_player = df.loc[df['Runs'].idxmax()]
print(Top_player)
```

```python
#Visualization:
# Pie Chart
plt.figure(figsize=(16, 4))
milestones = {"Centuries": 1, "Half-Centuries": 13, "Dots": 2}
plt.pie(milestones.values(), labels=milestones.keys(), autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightcoral'])
plt.title('Milestones Distribution')
plt.tight_layout()
plt.show()


```

```python
#Top 5 players by runs
Top_players = df.nlargest(5, 'Runs')
print(Top_players)
```

```python
#Ploting runs by players

Top_players_runs = Top_players.sort_values(by='Runs', ascending=False)
Top_players_sr = Top_players.sort_values(by='SR', ascending=False)
Top_players_50 = Top_players.sort_values(by='50', ascending=False)
fig, axes = plt.subplots(1, 3, figsize =(16, 4))

Top_players_runs.plot(x = 'Player', y = 'Runs', kind = 'bar', ax = axes[0])
axes[0].set_title('Runs by Players')
axes[0].tick_params(axis = 'x' ,rotation = 45)

#Top 5 players by SR
Top_players_sr.plot(x = 'Player', y = 'SR' , kind = 'bar', ax = axes[1])
axes[1].set_title('Top 5 Players by SR')
axes[1].tick_params(axis = 'x', rotation= 45)

# Top Players with most 50s
Top_players_50.plot(x = 'Player', y = '50', kind = 'bar', ax = axes[2])
axes[2].set_title('50 Runs by Players')
axes[2].tick_params(axis = 'x', rotation = 45)

plt.tight_layout()
plt.show()
```

```python
# List of player names to extract
players_to_extract = [ 'Q de Kock' , 'Aasif Sheikh', 'Aarif Sheikh', 'Aaron Jones', 'Fazalhaq Farooqi']

# Filter the data
filtered_data = df[df['Player'].isin(players_to_extract)]

print(filtered_data)
```

```python
#Visualization:
fig, axes = plt.subplots(1, 3, figsize = (16, 4))

filtered_data.plot(x = 'Player', y = 'SR' , kind = 'bar', ax = axes[0])
axes[0].set_title('SR of Players')
axes[0].tick_params(axis = 'x' ,rotation = 45)

filtered_data.plot(x = 'Player', y = 'Ave' , kind = 'bar', ax = axes[1])
axes[1].set_title('Average by Players')
axes[1].tick_params(axis = 'x' ,rotation = 45)

plt.bar(filtered_data['Player'], filtered_data['Runs'])
axes[2].set_title('Runs by Players')
axes[2].tick_params(axis = 'x' ,rotation = 45)

plt.tight_layout()
plt.show()
```
